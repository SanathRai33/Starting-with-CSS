#Starting with CSS

1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required

Ans-->
Index.html

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Priority Example</title>
    <!-- Internal CSS -->
    <style>
        div {
            background-color: skyblue;
            color: white;
            padding: 20px;
            text-align: center;
        }
    </style>
    <!-- External CSS -->
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <div style="background-color: red; border: 2px solid yellow;">
        <!-- Inline CSS (Highest Priority) -->
        This div is styled using inline, internal, and external CSS.
    </div>
    <div>
        <!-- External CSS (More Priority then Internal) -->
        This div is styled using internal, and external CSS.
    </div>
</body>
</html>


style.css

div {
    background-color: skyblue; /* External CSS */
    border: 2px solid black;
}

2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes

Ans-->
index.html

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BEM Naming Convention Example</title>
    <!-- External CSS -->
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <div class="text-block">
        <p class="text-block__paragraph">This is a normal paragraph.</p>
        <p class="text-block__paragraph text-block__paragraph--highlight">
            This is a highlighted paragraph.
        </p>
        <p class="text-block__paragraph text-block__paragraph--italic">
            This is an italicized paragraph.
        </p>
        <p class="text-block__paragraph text-block__paragraph--large">
            This is a larger paragraph.
        </p>
    </div>
</body>
</html>


style.css

/* Base styles */
.text-block__paragraph {
    font-size: 16px;
    color: #333;
    margin-bottom: 10px;
}

/* Modifier: Highlighted */
.text-block__paragraph--highlight {
    background-color: yellow;
    font-weight: bold;
}

/* Modifier: Italic */
.text-block__paragraph--italic {
    font-style: italic;
    color: gray;
}

/* Modifier: Large text */
.text-block__paragraph--large {
    font-size: 22px;
    color: blue;
}

3. Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

Ans-->
index.html

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled Form</title>
    <link rel="stylesheet" href="style.css"> <!-- External CSS -->
</head>
<body>
    <div class="form-container">
        <h2>Contact Form</h2>
        <form>
            <div class="form-section form-section--faded">
                <label for="name">Name:</label>
                <input type="text" id="name" name="name" placeholder="Enter your name">
                <label for="email">Email:</label>
                <input type="email" id="email" name="email" placeholder="Enter your email">
            </div>
            <div class="form-section">
                <label for="gender">Gender:</label>
                <select id="gender" name="gender">
                    <option value="male">Male</option>
                    <option value="female">Female</option>
                    <option value="other">Other</option>
                </select>
                <label for="message">Message:</label>
                <textarea id="message" name="message" rows="4" placeholder="Type your message..."></textarea>
                <button class="submit-button" type="submit">Submit</button>
            </div>
        </form>
    </div>
</body>
</html>


style.css

/* Custom color palette */
:root {
    --primary-color: #3498db;
    --secondary-color: #2ecc71;
    --background-color: #f4f4f4;
    --input-bg: #ffffff;
    --button-bg: #e74c3c;
    --text-color: #333;
}

body {
    font-family: Arial, sans-serif;
    background-color: var(--background-color);
    display: flex;
    justify-content: center;
    align-items: center;
    height: 100vh;
}

/* Form container */
.form-container {
    background: white;
    padding: 20px;
    border-radius: 8px;
    width: 350px;
    box-shadow: 0px 0px 10px rgba(0, 0, 0, 0.1);
}

/* Apply opacity to the first section */
.form-section--faded {
    opacity: 0.7;
}

/* Input field styling */
label {
    display: block;
    margin: 10px 0 5px;
    font-weight: bold;
    color: var(--text-color);
}

input, select, textarea {
    width: 100%;
    padding: 8px;
    margin-bottom: 10px;
    border: 1px solid #ccc;
    border-radius: 4px;
    background-color: var(--input-bg);
}

/* Focus effect */
input:focus, select:focus, textarea:focus {
    border-color: var(--primary-color);
    outline: none;
}

/* Button styling */
.submit-button {
    background: var(--button-bg);
    color: white;
    padding: 10px;
    border: none;
    width: 100%;
    cursor: pointer;
    font-size: 16px;
    border-radius: 4px;
}

.submit-button:hover {
    background: var(--secondary-color);
}
